# Data Pre-Processing

```python
Spotify_Most_Streamed_2024['Total Streams & Views'] = Spotify_Most_Streamed_2024['Spotify Streams'] + Spotify_Most_Streamed_2024['Pandora Streams'] + Spotify_Most_Streamed_2024['Soundcloud Streams'] + Spotify_Most_Streamed_2024['YouTube Views'] + Spotify_Most_Streamed_2024['TikTok Views']
Spotify_Most_Streamed_2024.head(2)
```

```python
Spotify_Most_Streamed_2024['Playlists'] = Spotify_Most_Streamed_2024['Spotify Playlist Count'] + Spotify_Most_Streamed_2024['Spotify Playlist Reach'] + Spotify_Most_Streamed_2024['YouTube Playlist Reach'] + Spotify_Most_Streamed_2024['Apple Music Playlist Count'] + Spotify_Most_Streamed_2024['Deezer Playlist Reach'] + Spotify_Most_Streamed_2024['Amazon Playlist Count']
Spotify_Most_Streamed_2024.head(2)
```

```python
Spotify_Most_Streamed_2024['Spotify'] = Spotify_Most_Streamed_2024['Spotify Streams'] + Spotify_Most_Streamed_2024['Spotify Playlist Count'] + Spotify_Most_Streamed_2024['Spotify Playlist Reach'] + Spotify_Most_Streamed_2024['Spotify Popularity']
Spotify_Most_Streamed_2024.head(2)
```

```python
Spotify_Most_Streamed_2024['YouTube'] = Spotify_Most_Streamed_2024['YouTube Views'] + Spotify_Most_Streamed_2024['YouTube Likes'] + Spotify_Most_Streamed_2024['YouTube Playlist Reach']
Spotify_Most_Streamed_2024.head(2)
```

```python
Spotify_Most_Streamed_2024['TikTok'] = Spotify_Most_Streamed_2024['TikTok Posts'] + Spotify_Most_Streamed_2024['TikTok Likes'] + Spotify_Most_Streamed_2024['TikTok Views']
Spotify_Most_Streamed_2024.head(2)
```

```python
Spotify_Most_Streamed_2024['Apple Music'] = Spotify_Most_Streamed_2024['Apple Music Playlist Count']
Spotify_Most_Streamed_2024.head(2)
```

```python
Spotify_Most_Streamed_2024['Deezer'] = Spotify_Most_Streamed_2024['Deezer Playlist Count'] + Spotify_Most_Streamed_2024['Deezer Playlist Reach']
Spotify_Most_Streamed_2024.head(2)
```

```python
Spotify_Most_Streamed_2024['Pandora'] = Spotify_Most_Streamed_2024['Pandora Streams'] + Spotify_Most_Streamed_2024['Pandora Track Stations']
Spotify_Most_Streamed_2024.head(2)
```

```python
Spotify_Most_Streamed_2024['Other'] = Spotify_Most_Streamed_2024['AirPlay Spins'] + Spotify_Most_Streamed_2024['SiriusXM Spins'] + Spotify_Most_Streamed_2024['Amazon Playlist Count'] + Spotify_Most_Streamed_2024['Shazam Counts'] + Spotify_Most_Streamed_2024['TIDAL Popularity']
Spotify_Most_Streamed_2024.head(2)
```
